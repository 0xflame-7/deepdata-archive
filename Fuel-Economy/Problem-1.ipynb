```python
import pandas as pd
import numpy as np
import tabulate 
import io
import os
```

```python
file_path = "./Data/"

for file in os.listdir(file_path):
	print(file)

df_list = [pd.read_excel(file_path + file,
					engine="xlrd" if file.endswith(".xls") else "openpyxl")
				  for file in os.listdir(file_path)]
```

```python
def create_table(task, header, data: list = df_list, start_year: int = 2008, table_design='plain', vertical=False, expand=False):
  table = []
  for i, df in enumerate(data, start=start_year):
    store = task(df)
    if expand:
      table.append([f"df_{i}", *store])
    else:
      table.append([f"df_{i}", store])
  
  if not vertical:
    table = list(map(list, zip(*([header] + table))))
    header = []
  
  return tabulate.tabulate(tabular_data=table, headers=header, tablefmt=table_design)
```

#### Basic Exploration

Number of Sample in each year

```python
def get_sample(data: pd.DataFrame):
  """Returns the number of samples in a DataFrame."""
  return data.size

print(create_table(header=["Dataset", "Number of Samples"], table_design="grid", task=get_sample))
```

Number of Columns in each dataset

```python
def get_shape(data: pd.DataFrame):
  """Returns the number of columns in a DataFrame."""
  return [data.shape[0], data.shape[1]]

print(create_table(header=["Dataset", "Number of Rows", "Number of Columns"], table_design="grid", task=get_shape, expand=True))
```

Columns

```python
def get_columns_value(data: pd.DataFrame):
  """Return list of all Columns"""
  Columns_value = data.columns.values
  returner = ''
  for value in Columns_value:
    returner += f"{value:<20}\n"
  return returner

print(create_table(task=get_columns_value, header=['Dataset', 'Columns'], table_design='grid'))
```

Duplicate rows in each dataset

```python
def get_duplicate(data: pd.DataFrame):
  return data.duplicated().sum()

print(create_table(task=get_duplicate, header=['Dataset', 'Number of Duplicate'], table_design='grid'))
```

Datatype in Columns

```python
def get_info(data: pd.DataFrame):
  buffer = io.StringIO()
  data.info(buf=buffer)
  return buffer.getvalue()

print(create_table(task=get_info, header=['Dataset', 'Info'], table_design='grid'))
```

Check Null

```python
def get_total_null(data: pd.DataFrame):
  return data.isnull().sum().sum()

print(create_table(task=get_total_null, header=['Dataset', 'Missing Values'], table_design="grid"))
```

```python
def get_null_value(data: pd.DataFrame):
  null_value_counts = data.isnull().sum()
  returner = ''
  for left, right in zip(null_value_counts.index, null_value_counts.values):
    returner += f'{left:<21}{right:>10}\n'
  return returner

print(create_table(task=get_null_value, header=["Dataset", "Missing Value"], table_design="grid"))
```

Number of Non-Null Unique values for features in each dataset

```python
def get_nunique(data: pd.DataFrame):
  Nunique_value = data.nunique()
  returner = ''
  for left, right in zip(Nunique_value.index, Nunique_value.values):
    returner += f'{left:<21}{right:>10}\n'
  return returner

print(create_table(task=get_nunique, header=['Dataset', 'Number of Non-Null Unique'], table_design="grid"))
```

Split(value/value) => (value, value)

```python
def spliter(data: pd.DataFrame, list_two_value_feature: list):
  df_1 = data.copy()
  df_2 = data.copy()

  for features in list_two_value_feature:
    df_1[features] = df_1[features].astype(str).apply(lambda x: x.split('/')[0])
    df_2[features] = df_2[features].astype(str).apply(lambda x: x.split("/")[1] if "/" in x else x)

  return df_1, df_2
```

```python
list_two_value_features = ['Fuel', 'Air_Pollution_Score', 'City_Mpg', 'Hwy_Mpg', 'Combined_Mpg', 'Greenhouse_Gas_Score']
Extra = 'Combined_Co2'

for i, data in enumerate(df_list):
  selected_data = data[data['Fuel'].astype(str).str.contains('/')]

  if selected_data.empty:
    continue

  # Dynamically add 'Combined_Co2' only if it exists in the current dataframe
  features_to_split = list_two_value_features.copy()
  if Extra in selected_data.columns:
    features_to_split.append(Extra)

  new_df_1, new_df_2 = spliter(selected_data, list_two_value_feature=features_to_split)

  data.drop(selected_data.index, inplace=True)
  df_list[i] = pd.concat([data, new_df_1, new_df_2], ignore_index=True)
```

Handling Fuel Data Inconsistencies

```python
def get_fuel(data: pd.DataFrame):
  Fuel_count = data['Fuel'].value_counts()
  returner = ''
  for fuel, value in Fuel_count.items():
    returner += f'{fuel:<20}{value:>10}\n'
  return returner

def get_smartway(data: pd.DataFrame):
	Count = data['Smartway'].value_counts()
	returner = ''
	for value, count in Count.items():
		returner += f'{value:<10}{count:>10}\n'
	return returner


print(create_table(task=get_fuel, header=['Dataset', 'Fuel'], table_design='grid'))
print(create_table(task=get_smartway, header=['Dataset', 'Smartway'], table_design='grid'))
```

```python
fuel_mapping = {
  "gasoline": "Gasoline",
  "ethanol": "Ethanol",
  "gas": "Gas",
  "diesel": "Diesel",
  "cng": "CNG",
  'electricity': "Electricity",
  "electricty": "Electricity",
  "hydrogen": "Hydrogen"
}


def rename_data(data):
  data['Fuel'] = data['Fuel'].str.lower().map(fuel_mapping)
  data['Smartway'] = data['Smartway'].apply(lambda x: x.capitalize())
  return data

for data in df_list:
  data = rename_data(data)

print(create_table(task=get_fuel, header=['Dataset', 'Fuel'], table_design='grid'))
print(create_table(task=get_smartway, header=['Dataset', 'Smartway'], table_design='grid'))
print(create_table(task=get_total_null, header=['Dataset', 'Number of Missing Value'], table_design='grid'))
```

Fix Mod (Air Pollution Score & Green House Gas Score)

```python
def get_air_pollution_score(data: pd.DataFrame):
  Count = data['Air_Pollution_Score'].value_counts()
  returner = ''
  for Score, count in Count.items():
    returner += f'{Score:<5}{count:>7}\n'
  return returner

def get_greenhouse_gas_score(data: pd.DataFrame):
  Count = data['Greenhouse_Gas_Score'].value_counts()
  returner = ''
  for Score, count in Count.items():
    returner += f'{Score:<5}{count:>7}\n'
  return returner



print(create_table(task=get_air_pollution_score, header=['Dataset', 'Air Pollution Score'], table_design="grid"))
print(create_table(task=get_greenhouse_gas_score, header=['Dataset', 'Greenhouse Gas Score'], table_design="grid"))
```

```python
# Replace Mod with 4
def replace_mod(score):
  if score == 'Mod':
    return '4'

  try:
    return score
  except ValueError:
    print(f"Invalid value: {score}")
    return np.nan

for data in df_list:
  data['Air_Pollution_Score'] = data['Air_Pollution_Score'].apply(replace_mod).astype(float)
  data['Greenhouse_Gas_Score'] = data[ 'Greenhouse_Gas_Score'].apply(replace_mod).astype('int64')


print(create_table(task=get_air_pollution_score, header=['Dataset', 'Air Pollution Score'], table_design="grid"))
print(create_table(task=get_greenhouse_gas_score, header=['Dataset', 'Greenhouse Gas Score'], table_design="grid"))
```

Fix Data type (City Miles per gallon, Highway Miles per gallon, Combined Miles per gallon)



```python
def get_city_miles_per_gallon(data: pd.DataFrame):
  Count = data['City_Mpg'].value_counts().head(10)
  returner = ''
  for Score, count in Count.items():
    returner += f'{Score:<5}{count:>7}\n'
  return returner

def get_Highway_miles_per_gallon(data: pd.DataFrame):
  Count = data['Hwy_Mpg'].value_counts().head(10)
  returner = ''
  for Score, count in Count.items():
    returner += f'{Score:<5}{count:>7}\n'
  return returner

def get_combined_mpg(data: pd.DataFrame):
  Count = data['Combined_Mpg'].value_counts().head(10)
  returner = ''
  for Score, count in Count.items():
    returner += f'{Score:<5}{count:>7}\n'
  return returner


print(create_table(task=get_city_miles_per_gallon, header=["Dataset", "City Miles Per Gallon"], table_design="grid"))
print(create_table(task=get_Highway_miles_per_gallon, header=["Dataset", "Highway Miles Per Gallon"], table_design="grid"))
print(create_table(task=get_combined_mpg, header=["Dataset", "Combined Miles per Gallon"], table_design="grid"))
```

Add Combined_Co2 (on 2008-2012)

![Combined_CO2](./img/CO2.png)

```python
fuel_emissions_map = {
    'Gasoline': {'co2_g_per_gal': 8887, 'ch4_n2o_g_per_mile': 1.9},
    'Ethanol': {'co2_g_per_gal': 6295, 'ch4_n2o_g_per_mile': 4.06},  # Mapping E85 to Ethanol
    'Gas': {'co2_g_per_gal': 7030, 'ch4_n2o_g_per_mile': 6.88},      # Assuming CNG
    'Diesel': {'co2_g_per_gal': 10180, 'ch4_n2o_g_per_mile': 0.4},
    'CNG': {'co2_g_per_gal': 7030, 'ch4_n2o_g_per_mile': 6.88},
    'Electricity': {'co2_g_per_gal': 0, 'ch4_n2o_g_per_mile': 0},    # Placeholder, since tailpipe = 0
}

def compute_comb_co2e(fuel_type: str, combined_mpg: float) -> float:
  fuel_type = fuel_type.strip()

  if fuel_type not in fuel_emissions_map:
    print(f"Unknown fuel type: '{fuel_type}'")
    raise ValueError(f"Unknown fuel type: '{fuel_type}'")

  if combined_mpg <= 0:
    print("Combined MPG must be a positive number.")
    raise ValueError("Combined MPG must be a positive number.")

  emission = fuel_emissions_map[fuel_type]
  
  co2_per_mile = emission['co2_g_per_gal'] / combined_mpg
  total_co2e = co2_per_mile + emission['ch4_n2o_g_per_mile']
  
  return round(total_co2e)


for data in df_list[:5]:
  def safe_compute(row):
    try:
      return compute_comb_co2e(row['Fuel'], float(row['Combined_Mpg']))
    except Exception:
      return None  # or np.nan

  data['Combined_Co2'] = data.apply(safe_compute, axis=1)
```
