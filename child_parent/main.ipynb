```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the dataset
file_path = "./parent_child_heights_1million.csv"  # Adjust path if needed
data = pd.read_csv(file_path)

# Parameters
mu = 100  # Population mean for height (child height)
alpha = 0.5  # Regression coefficient

# Step 2: Compute the regression formula for child heights (using the parent heights)
parent_heights = data['Parent Height (X)']
child_heights = mu + alpha * parent_heights + np.random.normal(0, 5, len(parent_heights))  # Adding noise

# Step 3: Calculate mean of parent and child heights
mean_parent_height = np.mean(parent_heights)
mean_child_height = np.mean(child_heights)

# Show Mean values
print(f"Mean Parent Height: {mean_parent_height:.2f}")
print(f"Mean Child Height: {mean_child_height:.2f}")

# Step 4: Extract extreme values (first and last rows)
extreme_values = data.iloc[[0, -1]]  # First and last (extreme) values
first_parent = extreme_values['Parent Height (X)'].iloc[0]
last_parent = extreme_values['Parent Height (X)'].iloc[-1]

first_child = mu + alpha * first_parent + np.random.normal(0, 5)
last_child = mu + alpha * last_parent + np.random.normal(0, 5)

# Show the extreme values and how they regress
print(f"\nExtreme Parent-Child Pairs:")
print(f"First Parent Height: {first_parent:.2f}, First Child Height: {first_child:.2f}")
print(f"Last Parent Height: {last_parent:.2f}, Last Child Height: {last_child:.2f}")

# Step 5: Plot the data to visualize the regression toward the mean
plt.figure(figsize=(10, 6))

# Scatter plot of Parent vs Child heights
plt.scatter(parent_heights, child_heights, alpha=0.5, color='blue', label="Data Points")
plt.title('Regression of Child Heights Toward the Mean Based on Parent Heights')
plt.xlabel('Parent Height (X)')
plt.ylabel('Child Height (Y)')
plt.grid(True)
plt.axhline(y=mu, color='r', linestyle='--', label=f'Mean Child Height (Y={mu})')  # Mean line for Child Height
plt.axvline(x=mean_parent_height, color='g', linestyle='--', label=f'Mean Parent Height (X={mean_parent_height:.2f})')
plt.legend(loc='upper left')
plt.show()
```
